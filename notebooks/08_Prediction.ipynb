```python
# ==========================================
# Fake News Detection Using NLP
# Phase 8 - Prediction System
# ==========================================

import re
import string
import joblib
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
model = joblib.load("../models/fake_news_model.pkl")

vectorizer = joblib.load("../models/tfidf_vectorizer.pkl")

print("Model Loaded Successfully.")
print("TF-IDF Vectorizer Loaded Successfully.")
```

```python
stop_words = set(stopwords.words("english"))

lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):

    text = text.lower()

    text = re.sub(r"http\S+|www\S+", "", text)

    text = re.sub(r"<.*?>", "", text)

    text = text.translate(
        str.maketrans("", "", string.punctuation)
    )

    text = re.sub(r"\d+", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    words = text.split()

    words = [
        word
        for word in words
        if word not in stop_words
    ]

    words = [
        lemmatizer.lemmatize(word)
        for word in words
    ]

    return " ".join(words)
```

```python
def predict_news(news_text):

    processed = preprocess_text(news_text)

    vector = vectorizer.transform([processed])

    prediction = model.predict(vector)[0]

    probability = model.predict_proba(vector)

    confidence = probability.max() * 100

    if prediction == 0:
        label = "🛑 Fake News"
    else:
        label = "✅ Real News"

    return label, confidence
```

```python
sample_news = """
President signs a new economic policy
to support small businesses across the country.
"""

print(predict_news(sample_news))
```

```python
print("=" * 60)
print("Fake News Detection System")
print("=" * 60)

news = input("Paste News Article:\n\n")

label, confidence = predict_news(news)

print()

print("=" * 60)

print("Prediction Result")

print("=" * 60)

print("Prediction :", label)

print(f"Confidence : {confidence:.2f}%")
```

```python
samples = [

"""NASA successfully launched a satellite into orbit.""",

"""Scientists confirmed the moon is made of cheese.""",

"""The government announced a new education policy."""
]

for i, article in enumerate(samples, 1):

    label, confidence = predict_news(article)

    print("=" * 50)

    print("Article", i)

    print(label)

    print(f"Confidence : {confidence:.2f}%")
```

```python
results = []

for article in samples:

    label, confidence = predict_news(article)

    results.append({

        "News": article,

        "Prediction": label,

        "Confidence (%)": round(confidence,2)

    })

import pandas as pd

prediction_df = pd.DataFrame(results)

prediction_df
```

```python
prediction_df.to_csv(

    "../reports/sample_predictions.csv",

    index=False

)

print("Prediction Report Saved Successfully.")
```
