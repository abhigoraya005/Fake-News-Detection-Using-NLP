```python
# ==========================================
# Fake News Detection Using NLP
# Phase 5 - Feature Engineering (TF-IDF)
# ==========================================

import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import joblib
import os
```

```python
news = pd.read_csv("../dataset/cleaned/clean_news.csv")

print("="*60)
print("Clean Dataset")
print("="*60)

print(news.head())

print()

print("Shape :", news.shape)
```

```python
print("="*60)
print("Missing Values")
print("="*60)

print(news.isnull().sum())
```

```python
X = news["text"]

y = news["label"]

print("Total Samples :", len(X))
print("Labels :", y.unique())
```

```python
vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2),
    stop_words="english"
)
```

```python
X_tfidf = vectorizer.fit_transform(X)

print("="*60)
print("TF-IDF Matrix")
print("="*60)

print(X_tfidf)

print()

print("Shape :", X_tfidf.shape)
```

```python
import joblib
import os

os.makedirs("../models", exist_ok=True)

joblib.dump(vectorizer, "../models/tfidf_vectorizer.pkl")

print("TF-IDF Vectorizer Saved Successfully.")
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X_tfidf,
    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)
```

```python
print("="*60)
print("Training and Testing Data")
print("="*60)

print("Training Samples :", X_train.shape[0])

print("Testing Samples :", X_test.shape[0])

print("Number of Features :", X_train.shape[1])
```

```python
import joblib
import os

os.makedirs("../models", exist_ok=True)

# Save TF-IDF Vectorizer
joblib.dump(vectorizer, "../models/tfidf_vectorizer.pkl")

print("TF-IDF Vectorizer Saved Successfully.")
```

```python
joblib.dump(X_train, "../models/X_train.pkl")

joblib.dump(X_test, "../models/X_test.pkl")

joblib.dump(y_train, "../models/y_train.pkl")

joblib.dump(y_test, "../models/y_test.pkl")

print("Training and Testing Data Saved Successfully.")
```

```python
feature_names = vectorizer.get_feature_names_out()

print("="*60)
print("Top 20 TF-IDF Features")
print("="*60)

print(feature_names[:20])
```

```python
print("="*60)
print("Feature Engineering Observations")
print("="*60)

print("1. Converted cleaned news articles into TF-IDF feature vectors.")
print("2. Used unigram and bigram features for better text representation.")
print("3. Limited the vocabulary to the top 5,000 informative terms.")
print("4. Split the dataset into 80% training and 20% testing samples.")
print("5. Saved the TF-IDF vectorizer for future predictions.")
print("6. Stored the training and testing datasets for machine learning models.")
print("7. The dataset is now ready for model training and evaluation.")

print("\nFeature Engineering Completed Successfully.")
```
