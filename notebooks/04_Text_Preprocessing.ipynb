```python
# ==========================================
# Fake News Detection Using NLP
# Phase 3 - Text Preprocessing
# ==========================================

import pandas as pd
import numpy as np
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required resources
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
news = pd.read_csv("../dataset/processed/news.csv")

print("="*60)
print("Processed Dataset")
print("="*60)

print(news.head())

print()

print("Shape :", news.shape)
```

```python
news = news[["text","label"]]

print(news.head())
```

```python
print("="*60)
print("Missing Values")
print("="*60)
print(news.isnull().sum())
```

```python
news.dropna(inplace=True)

news.reset_index(drop=True, inplace=True)

print(news.shape)
```

```python
news["text"] = news["text"].str.lower()

print(news["text"].head())
```

```python
news["text"] = news["text"].apply(
    lambda x: re.sub(r"http\S+|www\S+", "", x)
)
```

```python
news["text"] = news["text"].apply(
    lambda x: re.sub(r"<.*?>", "", x)
)
```

```python
def remove_punctuation(text):
    return text.translate(
        str.maketrans(
            "",
            "",
            string.punctuation
        )
    )

news["text"] = news["text"].apply(remove_punctuation)
```

```python
news["text"] = news["text"].apply(
    lambda x: re.sub(r"\d+", "", x)
)
```

```python
news["text"] = news["text"].apply(
    lambda x: re.sub(r"\s+", " ", x).strip()
)
```

```python
stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
    words = text.split()

    words = [
        word
        for word in words
        if word not in stop_words
    ]

    return " ".join(words)

news["text"] = news["text"].apply(remove_stopwords)
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):

    words = text.split()

    words = [
        lemmatizer.lemmatize(word)
        for word in words
    ]

    return " ".join(words)

news["text"] = news["text"].apply(lemmatize_text)
```

```python
print("="*60)
print("Cleaned News Articles")
print("="*60)

print(news.head())
```

```python
# Remove any rows with missing text
news = news.dropna(subset=["text"])

# Remove rows where text became empty
news = news[news["text"].str.strip() != ""]

# Reset index
news = news.reset_index(drop=True)

print(news.shape)
print(news.isnull().sum())
```

```python
import os

os.makedirs("../dataset/cleaned", exist_ok=True)

news.to_csv(
    "../dataset/cleaned/clean_news.csv",
    index=False
)

print("Clean Dataset Saved Successfully.")
```
