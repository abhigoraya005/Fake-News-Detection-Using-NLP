```python
  # ==========================================
# Fake News Detection using NLP
# Phase 2 - Data Understanding
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")      
```

```python
fake = pd.read_csv("../dataset/raw/Fake.csv")
true = pd.read_csv("../dataset/raw/True.csv")

print("="*60)
print("Fake News Dataset")
print("="*60)
print(fake.head())

print()

print("="*60)
print("True News Dataset")
print("="*60)
print(true.head())
```

```python
print("="*60)
print("Fake Dataset Information")
print("="*60)
fake.info()

print()

print("="*60)
print("True Dataset Information")
print("="*60)
true.info()
```

```python
print("="*60)
print("Dataset Shape")
print("="*60)

print("Fake Dataset :", fake.shape)
print("True Dataset :", true.shape)

print()

print("="*60)
print("Columns")
print("="*60)

print(fake.columns.tolist())
```

```python
print("="*60)
print("Missing Values - Fake Dataset")
print("="*60)

print(fake.isnull().sum())

print()

print("="*60)
print("Missing Values - True Dataset")
print("="*60)

print(true.isnull().sum())
```

```python
print("="*60)
print("Duplicate Records")
print("="*60)

print("Fake :", fake.duplicated().sum())
print("True :", true.duplicated().sum())
```

```python
fake["label"] = 0
true["label"] = 1

print(fake.head())

print()

print(true.head())
```

```python
news = pd.concat([fake, true], ignore_index=True)

print("="*60)
print("Combined Dataset")
print("="*60)

print(news.head())

print()

print("Combined Shape :", news.shape)
```

```python
news = news.sample(frac=1, random_state=42)

news.reset_index(drop=True, inplace=True)

print(news.head())
```

```python
print("="*60)
print("Dataset Statistics")
print("="*60)

print(news.describe(include="all"))
```

```python
news.to_csv("../dataset/processed/news.csv", index=False)

print("Processed Dataset Saved Successfully.")
```
